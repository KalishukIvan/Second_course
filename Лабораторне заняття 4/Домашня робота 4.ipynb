## 1 Від'ємний біноміальний розподіл

Більш детальна інформація в [документації](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.nbinom.html).

```python
from scipy.stats import nbinom, binom, geom, poisson
```

```python
n = 5
p = 0.4
```

### 1.1 Щільність

```python
x = np.arange(-0.5, n*5+1, 0.5)
```

Виведіть значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
thickness = nbinom.pmf(x,n,p)
print(thickness)
```

Зобразіть графічно значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
plt.stem(x,thickness)
plt.show()
```

Зобразіть графічно значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x1$.

```python
x1 = np.arange(0, n*5+1)
```

```python
thickness = nbinom.pmf(x1,n,p)
plt.stem(x1,thickness)
plt.show()
```

### 1.2 Функція розподілу

Виведіть значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
rosp = nbinom.cdf(x,n,p)
print(rosp)
```

Зобразіть графічно значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
plt.plot(x, rosp, drawstyle='steps-post')
plt.show()
```

### 1.3 Випадкові величини

Згенеруйте $100$ випадкових величин з від'ємного біноміального розподілу з параметрами $n, p$.

```python
numbers = nbinom.rvs(n = n, p = p, size = 100)
print(numbers)
```

Виведіть кількість кожного елемента в отриманому випадковому векторі.

```python
for i in range(max(numbers)+1):
    tmp = sum(numbers == i)
    if tmp != 0:
        print("Number of {} -- {}".format(i,tmp))
```

Зобразіть графічно кількість кожного елемента в отриманому випадковому векторі.

```python
plt.hist(numbers)
plt.show()
```

## 3 Різні параметри

### 3.1 Біноміальний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин біноміального розподілу з параметрами:
1. $n = 15, p = 0.8$
2. $n = 15, p = 0.2$
3. $n = 20, p = 0.4$
4. $n = 20, p = 0.6$
5. $n = 7, p = 0.5$
6. $n = 25, p = 0.5$
7. $n = 100, p = 0.5$

```python
def task_3_1(data):
    f, axes = plt.subplots(3, 7, figsize = (60,30))
    for i,d in enumerate(data):
        n,p = d
        x = np.arange(-0.5, n + 1, 0.5)
        # print(x)
        thickness = binom.pmf(x, n, p)
        rosp = binom.cdf(x, n, p)
        numbers = binom.rvs(n=n, p=p, size=10000)
        axes[0,i].stem(x, thickness, label = 'n = {}, p = {}'.format(n,p))
        axes[1,i].plot(x, rosp, drawstyle='steps-post', label = 'n = {}, p = {}'.format(n,p))
        axes[2,i].hist(numbers, label = 'n = {}, p = {}'.format(n,p))
        for j in range(3):
            axes[j,i].legend(loc = 2)
    plt.show()

data = [(15, 0.8), (15, 0.2), (20, 0.4), (20, 0.6), (7, 0.5), (25, 0.5), (100, 0.5)]
task_3_1(data)

```

### 3.2 Геометричний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин геометричного розподілу з параметрами:

5. $p = 0.5$
6. $p = 0.6$
7. $p = 0.7$
8. $p = 0.8$
9. $p = 0.9$

```python
def task_3_2(data):
    f, axes = plt.subplots(3, 9, figsize=(50, 25))
    for i, p in enumerate(data):
        x = np.arange(0, 11, 0.5)
        thickness = geom.pmf(x, p)
        rosp = geom.cdf(x, p)
        numbers = geom.rvs(p=p, size=10000)
        axes[0, i].stem(x, thickness, label='p = {}'.format(p))
        axes[1, i].plot(x, rosp, drawstyle='steps-post', label='p = {}'.format(p))
        axes[2, i].hist(numbers, label='p = {}'.format(p))
        for j in range(3):
            axes[j, i].legend(loc=2)
    plt.show()
    
data = [i/10 for i in range(1,10)]
task_3_2(data)
```

### 3.3 Розподіл Пуассона

```python
def task_3_3(data):
    f, axes = plt.subplots(3, 9, figsize=(40, 20))
    for i, mu in enumerate(data):
        x = np.arange(-0.5, 11, 0.5)
        thickness = poisson.pmf(x, mu)
        rosp = poisson.cdf(x, mu)
        numbers = poisson.rvs(mu=mu, size=10000)
        axes[0, i].stem(x, thickness, label='mu = {}'.format(mu))
        axes[1, i].plot(x, rosp, drawstyle='steps-post', label='mu = {}'.format(mu))
        axes[2, i].hist(numbers, label='mu = {}'.format(mu))
        for j in range(3):
            axes[j, i].legend(loc=2)
    plt.show()
    
data = [i/10 for i in range(1,10)]
task_3_3(data)
```

## 4 Апроксимація біноміального розподілу розподілом Пуассона

Покажіть схожість біноміального розподілу та розподілу Пуассона (їх функцій щільності) з відповідними параметрами.
Наприклад:
- $n = 100, p = 0.05$
- $n = 100, p = 0.1$
- $n = 100, p = 0.15$
- $n = 100, p = 0.2$
- $n = 100, p = 0.35$
- $n = 100, p = 0.5$
- $n = 100, p = 0.75$

```python
def task_4(data):
    f, axes = plt.subplots(2, 7, figsize=(60, 30))
    for i, d in enumerate(data):
        n, p = d
        x = np.arange(-0.5, n + 1, 0.5)
        thickness_b = binom.pmf(x, n, p)
        thickness_g = geom.pmf(x, p)
        axes[0, i].stem(x, thickness_b, label='binom : n = {}, p = {}'.format(n,p))
        axes[1, i].stem(x, thickness_g, label='geom : p = {}'.format(p))
        
        for j in range(2):
            axes[j, i].legend(loc=2)
    plt.show()
    
data = [(100, 0.05), (100, 0.1), (100, 0.15), (100, 0.2), (100, 0.35), (100, 0.5), (100, 0.75)]
task_4(data)
```
